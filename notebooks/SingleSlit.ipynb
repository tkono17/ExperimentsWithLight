```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Parameters
wavelength = 635*1.0E-6 # 635 [nm] -> [mm]
b = 0.02 # 20 [um]
l = 300.0 # [mm]
x0 = 150.0 # [mm]
xmin = 130.0 # [mm]
A = 1.0 # Amplitude [mA]

N = 10000
dx = (x0 - xmin)*2/N
k = 2*math.pi/wavelength
x = xmin + np.array(range(N) )*dx
theta = np.atan( (x-x0)/l)
beta = k*b*np.sin(theta)
R = np.sqrt(l**2 + x**2)
I = A/(2*R) * (np.sin(beta)/beta)**2

plt.plot(x, I)
plt.show()
```
