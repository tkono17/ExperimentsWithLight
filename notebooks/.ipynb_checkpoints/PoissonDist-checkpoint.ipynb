```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

x = np.array(range(20) )

y1 = poisson.pmf(x, mu=1.0)

plt.plot(x, poisson.pmf(x, mu=0.2), marker='o', label='mu=0.2')
plt.plot(x, poisson.pmf(x, mu=1.0), marker='s', label='mu=1')
plt.plot(x, poisson.pmf(x, mu=2.0), marker='v', label='mu=2')
plt.plot(x, poisson.pmf(x, mu=5.0), marker='^', label='mu=5')
plt.plot(x, poisson.pmf(x, mu=10.0), marker='1', label='mu=10')
plt.legend()
plt.show()
```
