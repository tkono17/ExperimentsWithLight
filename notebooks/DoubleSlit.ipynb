```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Parameters
wavelength = 635*1.0E-6 # 635 [nm] -> [mm]
a = 0.05 # [mm]
b = 0.02 # 20 [um]
l = 300.0 # [mm]
x0 = 150.0 # [mm]
xmin = 130.0 # [mm]
A = 1.0 # Amplitude [mA]

def plotPattern(N=10000):
    dx = (x0 - xmin)*2/N
    k = 2*math.pi/wavelength
    x = xmin + np.array(range(N) )*dx
    theta = np.atan( (x-x0)/l)
    alpha = k*a*np.sin(theta)
    beta = k*b*np.sin(theta)
    R = np.sqrt(l**2 + x**2)
    Ib = x
    with np.errstate(all='ignore'):
        Ib = A/(2*R) * (np.sin(beta)/beta)**2
    I = Ib * np.cos(alpha)**2

    plt.plot(x, Ib, color='red', linestyle='dashed')
    plt.plot(x, I)

plotPattern()
plt.show()
```

```python
data_x = [130.0, 140, 150, 160, 170 ]
data_y = [0.0, 0.1, 1.4, 0.2, 0.0]

A = 1000.0 # amplitude

plt.scatter(data_x, data_y, marker='o', color='black')
plotPattern()
plt.show()
```
